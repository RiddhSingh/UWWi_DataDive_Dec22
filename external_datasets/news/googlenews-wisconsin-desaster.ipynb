```python
! pip install GoogleNews

from GoogleNews import GoogleNews

googlenews = GoogleNews(start='01/01/2017',end='11/30/2022')
googlenews.search('Wisconsin Desaster')
```

```python
type(googlenews)
```

```python
dir(googlenews)
```

```python
result = []

for i in range(1, 999):
    try:
        result.append(googlenews.page_at(i))
    except Exception as e:
        print(i, e)

print(len(result))
```

```python
print(len(result))
```

```python
final = []
for r in result[:12]:
    for el in r:
        final.append(el)
print(len(final))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(final)
```

```python
df.to_csv("wisconsin_desaster.csv")
```
